# Correlación de variables

- La **correlación** es una unidad de medida que resume la relación de distribución entre 2 variables. Que haya correlación no implica que haya *causalidad* entre las variables.
- Una forma de **cuantificar** esta relación es utilizando el **coeficiente de correlación de Pearson**, el cual es una medida de la *asociación lineal* entre 2 variables.
- El coeficiente de Pearson tiene un valor entre -1 y 1 <br>
 **-1** → *correlación lineal perfecta negativa:* cuando una variable aumenta la otra disminuye.<br>
 **0** → *no hay correlación lineal*: para cada comportamiento de la variable x, la variable y toma un valor indistinto. Nube de puntos.<br>
 **1 →** *correlación lineal perfecta positiva*: cuando una variable aumenta la otra también
 
 
### Objetivo:
- Determinar la correlación de las variables del dataset "penguins"

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('penguins-clean.csv')
data.head()
```

```python
data.corr(method='pearson')
```

```python
sns.heatmap(data.corr(method='pearson'), cmap="crest", linewidth=.7)
```

## Observaciones:
Los índices de correlación mas altos son:
<br>
- Una correlación de **0.87** entre las variables flipper_lenght_mm (largo de la aleta) y body_mass_g (masa corporal).
- Entre flipper_lenght_mm (largo de la aleta) y Species_Gentoo (especie Gentoo) el índice de correlación es de **0.86**
- Las variables Species_Gentoo (especie Gentoo) y body_mass_g (masa corporal) tienen una correlación de **0.82**
- Species_Gentoo (especie Gentoo) y Island_Biscoe (Isla Biscoe) tienen un indice de correlación de **0.76**
<br>

Las relaciones entre estas variables son las mas próximas a una relación lineal positiva, lo cual no implica que haya una relación de causa-efecto entre ellas.
